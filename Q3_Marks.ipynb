```python
# Function to manage marks

def marks():

    marks = []

    try:

        for i in range(5):
            m = float(input("Enter mark: "))
            marks.append(m)

        print("Marks =", marks)

        avg = sum(marks) / len(marks)
        print("Average =", avg)

        print("Highest =", max(marks))
        print("Lowest =", min(marks))

        marks.sort(reverse=True)
        print("Descending Order =", marks)

    except ValueError:
        print("Please enter numbers only.")

marks()
```
