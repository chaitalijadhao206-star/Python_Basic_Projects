```python
import math

try:

    s = input("Enter Sentence: ")

    words = s.split()

    u = set(words)

    print("Unique Words =")

    for i in sorted(u):
        print(i)

    n = len(u)

    print("Power =", math.pow(n, 2))

except:
    print("Something went wrong")
```
