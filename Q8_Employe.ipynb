```python
# Employee Class

class Employee:

    def __init__(self, id, name, dept, salary):
        self.id = id
        self.name = name
        self.details = (dept, salary)

    def show_details(self):
        print("ID =", self.id)
        print("Name =", self.name)
        print("Department =", self.details[0])
        print("Salary =", self.details[1])
        print()


d = {}

for i in range(3):

    id = input("Enter ID: ")
    name = input("Enter Name: ")
    dept = input("Enter Department: ")
    salary = int(input("Enter Salary: "))

    e = Employee(id, name, dept, salary)

    d[id] = e


print("\nEmployee Details\n")

for e in d.values():
    e.show_details()
```
