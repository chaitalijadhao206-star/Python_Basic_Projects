```python
# Student Database

def student_database():

    d = {}

    while True:

        print("\n1. Add Student")
        print("2. Search Student")
        print("3. Display All")
        print("4. Exit")

        try:
            ch = int(input("Enter Choice: "))

            if ch == 1:
                roll = input("Enter Roll No: ")
                name = input("Enter Name: ")
                age = int(input("Enter Age: "))
                city = input("Enter City: ")

                d.update({roll: [name, age, city]})
                print("Student Added")

            elif ch == 2:
                roll = input("Enter Roll No: ")
                s = d.get(roll)

                if s:
                    print("Name =", s[0])
                    print("Age =", s[1])
                    print("City =", s[2])
                else:
                    print("Student Not Found")

            elif ch == 3:
                for roll, s in d.items():
                    print("\nRoll =", roll)
                    print("Name =", s[0])
                    print("Age =", s[1])
                    print("City =", s[2])

            elif ch == 4:
                print("Thank You")
                break

            else:
                print("Invalid Choice")

        except ValueError:
            print("Enter Correct Input")


student_database()
```
