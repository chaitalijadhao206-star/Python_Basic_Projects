```python
def string(s):

    if s == "":
        print("Empty String")
        return

    print("Length =", len(s))
    print("Reverse =", s[::-1])

    count = 0

    for ch in s:
        if ch.lower() in "aeiou":
            count = count + 1

    print("Vowels =", count)

    print("Character with Index")

    for i in range(len(s)):
        print(i, i-len(s), s[i])


s = input("Enter String: ")

string(s)
```
