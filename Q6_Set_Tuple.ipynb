```python
import random
import math

def number():

    s = set()

    try:

        for i in range(10):
            n = int(input("Enter Number: "))
            s.add(n)

        print("Set =", s)

        t = tuple(s)
        print("Tuple =", t)

        r = random.sample(t, 3)
        print("Random Numbers =", r)

        total = sum(t)
        print("Square Root =", math.sqrt(total))

    except ValueError:
        print("Enter numbers only")

number()
```
