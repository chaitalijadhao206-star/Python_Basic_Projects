```python
# Lambda function

try:

    square = lambda x: x * x

    l = []

    for i in range(1, 21):
        l.append(square(i))

    print("Even Squares:")

    for n in l:
        if n % 2 == 0:
            print(n)

except Exception:
    print("Error")
```
